Problem 5 - Numlength 2 

```python
pos_int=[2,45,342,65,77,890,256,1995,12,2018,5,10,7]
```

```python
one_digit=[]
two_digit=[]
three_digit=[]
four_digit=[]

for i in pos_int:
    if 0<i<10:
        one_digit.append(i)
    elif 10<=i<100:
        two_digit.append(i)
    elif 100<=i<1000:
        three_digit.append(i)
    else:
        four_digit.append(i)
        
print("one_digit", one_digit)
print("two_digit", two_digit)
print("three_digit", three_digit)
print("four_digit", four_digit)
```

```python
prop_price=quandl.get('ZILLOW/C25499_MLPFAH',auth_token='bA2NtfQcjt2AYx1yoMJC')
```

```python
prop_price.head()
```

```python
condition_1=prop_price.Value>160
condition_2=prop_price.Value<170
len(prop_price[condition_1&condition_2])
```

```python
prop_price.hist(bins=15,color="b",alpha=0.7)
plt.xlabel("Values")
plt.ylabel("Name")
plt.title("Property Price")
```

```python
def classifier(x):
    if x>=170:
        outcome="High"
    elif 160<x<170:
        outcome="Medium"
    else:
        outcome="Low"
    return outcome
```

```python
prop_price["Classes"]=prop_price.Value.apply(classifier)
prop_price.head()
```

```python
pd.pivot_table(prop_price,values="Value",columns='Classes',aggfunc="std")
```
